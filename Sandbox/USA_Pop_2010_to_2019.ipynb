```python
# Dependencies
import pandas as pd
```

```python
# Import excel file into a data frame
excel_path = "Resources/USA_Resident_Population_2010_to_2019_N1.xlsx"
population_df1=pd.read_excel(excel_path)
population_df1.head()
```

```python
# Set data frame index to Country, County, State
population_df2=population_df1.set_index(['Country', 'County', 'State'])
population_df2
```

```python
# The rest of the columns will get an axis name "Year"
population_df2.rename_axis(columns=["Year"], inplace=True)
population_df2
```

```python
# Use the Stack fuction to pop out the year into a column
    # It becomes part of the multi-index
population_df3=pd.DataFrame(population_df2.stack())
population_df3
```

```python
# Renamed column name 0 to "Resident Population"
population_df3.rename(columns={0:"Resident Population"}, inplace=True)
population_df3
```

```python
population_df4=population_df3.reset_index()
population_df4
```

```python
# # Select for years 2010-2017
population_df5=population_df4.loc[(population_df4['Year']>=2010) & (population_df4['Year']<=2017)]
population_df5
```

```python
# Copy and Display data frame
population_df=population_df5.copy()
population_df
```
