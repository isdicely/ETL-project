```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import csv file into a data frame
csv_path = "Resources/bestsellers with categories.csv"
best_sellers_df1=pd.read_csv(csv_path)
best_sellers_df1
```

```python
# Sort data frame by year (ascending) and set the index to the name of the best sellers
best_sellers_df2= best_sellers_df1.sort_values(by=['Year']).set_index("Name")
```

```python
# Display the data frame
best_sellers_df2
```

```python
# Select for years 2010-2017
best_sellers_df3=best_sellers_df2.loc[(best_sellers_df2['Year']>=2010) & (best_sellers_df2['Year']<=2017)]
```

```python
# Display the data frame
best_sellers_df3
```

```python
#Connect to local database
connection_string = "postgres:password@localhost:5432/loading_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#check for tables
engine.table_names()
```

```python
#Use pandas to load csv converted DataFrame into database
best_sellers_df3.to_sql(name='best_sellers', con=engine, if_exists='append', index=True)
```

```python
#read the table back
pd.read_sql_query('select * from best_sellers', con=engine).head()
```
