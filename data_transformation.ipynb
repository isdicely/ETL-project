```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Import csv file into a data frame
csv_path = "Resources/bestsellers with categories.csv"
best_sellers_df1=pd.read_csv(csv_path)
best_sellers_df1
```

```python
# Sort data frame by year (ascending) and set the index to the name of the best sellers
best_sellers_df2= best_sellers_df1.sort_values(by=['Year']).set_index("Name")
```

```python
# Display the data frame
best_sellers_df2
```

```python
# Select for years 2010-2017
best_sellers_df3=best_sellers_df2.loc[(best_sellers_df2['Year']>=2010) & (best_sellers_df2['Year']<=2017)]
```

```python
# Copy and Display the data frame 
best_sellers_df=best_sellers_df3.copy()
```

```python
# Reading in Excel Sheet
la_file = "Resources/living_arrangements.xlsx.xls"
living_arr = pd.read_excel(la_file)

# Dropping Null Values
c_la=living_arr.dropna(how='any')

# Resetting Column Headers
headers = c_la.iloc[0]
new_df  = pd.DataFrame(c_la.values[0:], columns=headers)
new_df_c= new_df.drop([0])

# Dropping unnecessary years
c_adl=new_df_c.loc[(new_df_c['Year']>=2010) & (new_df_c['Year']<=2017)]
c_adl
```

```python
# Reading in excel sheet
yb_file = "Resources/yearly_births.xlsx"
yearly_b = pd.read_excel(yb_file)

# Dropping Null Values
c_yb=yearly_b.dropna(how='any')

# Renaming column headers
cl_yb=c_yb.rename({'Table with row headers in column A and column headers in rows ^ and 7.':'Year',
                   'Unnamed: 1':'15 to 50','Unnamed: 2':'Margin of error/1','Unnamed: 3':'15 to 19',
                   'Unnamed: 4':'Margin of error /1','Unnamed: 5':'20 to 24','Unnamed: 6':'Margin of error /1',
                   'Unnamed: 7':'25 to 29','Unnamed: 8':'Margin of error /1','Unnamed: 9':'30 to 34',
                   'Unnamed: 10':'Margin of error /1','Unnamed: 11':'35 to 39','Unnamed: 12':'Margin of error /1',
                   'Unnamed: 13':'40 to 44','Unnamed: 14':'Margin of error /1','Unnamed: 15': '45 to 55',
                   'Unnamed: 16': 'Margin of error /1'}, axis='columns')

yearly_birth=cl_yb.drop([15,16,17,18])
yearly_birth
```

```python
# Import excel file into a data frame
excel_path = "Resources/USA_Resident_Population_2010_to_2019_N1.xlsx"
population_df1=pd.read_excel(excel_path)
population_df1.head()
```

```python
# Set data frame index to Country, County, State
population_df2=population_df1.set_index(['Country', 'County', 'State'])
population_df2
```

```python
# The rest of the columns will get an axis name "Year"
population_df2.rename_axis(columns=["Year"], inplace=True)
population_df2
```

```python
# Use the Stack fuction to pop out the year into a column
    # It becomes part of the multi-index
population_df3=pd.DataFrame(population_df2.stack())
population_df3
```

```python
# Renamed column name 0 to "Resident Population"
population_df3.rename(columns={0:"Resident Population"}, inplace=True)
population_df3
```

```python
population_df4=population_df3.reset_index()
population_df4
```

```python
# # Select for years 2010-2017
population_df5=population_df4.loc[(population_df4['Year']>=2010) & (population_df4['Year']<=2017)]
population_df5
```

```python
# Copy and Display data frame
population_df=population_df5.copy()
population_df
```

```python
#Connect to local database
connection_string = "postgres:password@localhost:5432/loading_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#check for tables
engine.table_names()
```

```python
#Use pandas to load csv converted DataFrame into database
best_sellers_df.to_sql(name='best_sellers', con=engine, if_exists='append', index=True)
c_adl.to_sql(name='adult_housing', con=engine, if_exists='append', index=True)
yearly_birth.to_sql(name='us_births', con=engine, if_exists='append', index=True)
population_df.to_sql(name='us_population', con=engine, if_exists='append', index=True)
```

```python
#read the table back
pd.read_sql_query('select * from best_sellers', con=engine).head()
```
